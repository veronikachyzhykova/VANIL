***Chyzhykova***

```python
'''1
кількість червоних та жовтих карток при спаданню вдалої передачі м'яча,
тобто бачимо, що існує тенденція залежності штрафних карток від високого рівня передачі'''


i = data.loc[:, ['Team','Interceptions','Yellow Cards', 'Red Cards']].sort_values(by = 'Interceptions', ascending = False)
i['Sum Cards']=i['Yellow Cards']+i['Red Cards']
i
```

```python
'''2
команди, котрі зіграли погано відносно голів та попадання повз ціль'''

bad = data.loc[(data['Goals'] < 8 )&(data['Shots off target']>20)]
bad.loc[:,['Team','Goals','Shots off target','% Goals-to-shots', 'Red Cards', 'Yellow Cards']].sort_values(by = 'Goals',ascending = False )
```

```python
'''3
команди, що забивали не класичні голи, принаймні 1 гол головою наприклад'''

creative_goals = data[data['Headed goals'] != 0].loc[:,['Team','Penalty goals','Headed goals']].sort_values(by = ['Penalty goals','Headed goals'],ascending = False )
creative_goals
```

***Khariton***

```python
'''4
Команди що мали точніть ударів більше 45% їх кількість голів,назви,точність ударів в порядку спадання поголам
тобто розуміємо, що висока точність ударів не гарантує високу к-сть голів'''

shots = data[data['Shooting Accuracy'] > (str(45)+'%')][['Team','Goals','Shooting Accuracy']].sort_values('Goals', ascending=False)

shots
```

```python
'''5
Скільки червоних карток малик команди, що мали більше 10 жовтих карток за порушення відсортовані в порядку спадання
'''

cards = data[data['Yellow Cards'] > 10 ][['Team','Yellow Cards','Red Cards']].sort_values('Yellow Cards', ascending=False)

cards
```

```python
'''6
Команди що забили більлше 5голів за чемпіонат 
кількість голів / кількіть точних ударів в ворота / %голів від ударів відсортовано в порядку спадання'''

teams = pd.DataFrame()
teams = data[data['Goals'] > 5 ][['Team','Shots on target','Goals']]
teams['%'] = teams['Goals'] / teams['Shots on target'] * 100
teams.sort_values('%',ascending = False)

teams
```

***Chemerys***

```python
'''7
Яка команда грала краще в обороні? (сухі матчі, блоки, виноси, перехоплення, відбори)'''

protect = data.loc[:,['Team','Clean Sheets','Blocks','Clearances','Interceptions','Tackles']].loc[(data['Clean Sheets']>=1)&(data['Blocks']>11)&(data['Clearances']>=90)&(data['Interceptions']>=70)&(data['Tackles']>=50)].sort_values('Clean Sheets', ascending=False).head(1)
protect
```

***Mucha***

```python
'''8
наскільки залежить вдала гра голкіпера(не пропускаючи) від захисників команди'''

playing_defense = data.loc[:, ['Team', 'Clean Sheets', 'Clearances off line', 'Blocks', 'Interceptions']].sort_values(by = 'Clean Sheets', ascending = False)
playing_defense['the number of successful actions in defense'] = playing_defense['Clearances off line']+playing_defense['Blocks']+playing_defense['Interceptions']
playing_defense

```

```python
'''9
топ 5 команд, що найбільше порушували правила, в таблиці, за червона карта, жовта, офсайд, фол'''

the_biggest_offenders = pd.DataFrame()
the_biggest_offenders = data.loc[:,['Team', 'Red Cards', 'Yellow Cards', 'Offsides', 'Fouls Conceded']]
the_biggest_offenders['assessment of violations'] = the_biggest_offenders['Red Cards'] + the_biggest_offenders['Yellow Cards'] + the_biggest_offenders['Offsides'] + the_biggest_offenders['Fouls Conceded']

the_biggest_offenders.sort_values('assessment of violations', ascending = False).head(5)


```

```python
'''10
Команди, де % забитих м'ячів до кількості гравців, які зіграли, більше, ніж 50%'''

goalPlayer = pd.DataFrame()
goalPlayer = data.loc[:,['Team', 'Goals', 'Players Used']]
goalPlayer['%'] = goalPlayer['Goals']*100 / goalPlayer['Players Used']
goalPlayer.loc[(goalPlayer['%']>50)].sort_values('%', ascending = False)

```
