## 6.Напишіть програму Python, яка приймає ціле число (n) і обчислює значення n+nn+nnn

```python
n = int(input('Введіть ціле число: '))
S = n +(n*10+n)+(n*100+n*10+n)
print(S) 
```

## 15.Магазин бере 12 доларів за товар, якщо ви купуєте менше 10 предметів. Якщо ви купуєте від 10 до 99 предметів, вартість становить 10 доларів США за одиницю. Якщо ви купуєте 100 або більше предметів, вартість становить 7 доларів США за товар. Напишіть програму, яка запитує користувача, скільки товарів він купує, і друкує загальну вартість.

```python
x = int(input('Введіть кількість товарів для покупки: '))
if x < 10:
  x = x*12
elif 9 < x < 100 :
  x = x*10
else :
  x = x* 7
print(x) 
```

## Сортувати кортеж кортежів за другим елементом. Наприклад, вихід для кортежу ((′a′,23),(′b′,37),(′c′,11),(′d′,29)) має бути ((′c′,11) ),(′a′,23),(′d′,29),(′b′,37))

```python
from operator import itemgetter

list_of_tuples = [('a', 23),
('d', 37),
('c_', 11),
('d', 29)]

sorted(list_of_tuples, key=itemgetter(1))
```
