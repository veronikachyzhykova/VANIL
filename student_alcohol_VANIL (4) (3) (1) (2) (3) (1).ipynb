```python
'''1
Порівняємо середній вік дівчат та хлопців за рівнем вживання алкоголю у будні дні'''

age_alc = data.groupby(['Dalc','sex']).age.mean()
pd.DataFrame(age_alc)
```

```python
'''2
яка кількість та рівень здоров'я в учнів, що випивають у будні та вихідні дні не менше 1 рівня алкоголю
health	current health status (numeric: from 1 - very bad to 5 - very good)'''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
health = data.loc[(data['Mean_of_alc'] >= 1)].groupby(['Mean_of_alc']).health.describe()
health.sort_values(by = 'Mean_of_alc', ascending= False)
```

```python
'''3
пропорція вживання алкоголю учнями на рівень, що більший 2: повнолітні/неповнолітні за Конституцією України '''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
adults = data[(data.age >= 18)&(data.Mean_of_alc >2)].sex.count()
children = data[(data.age < 18)&(data.Mean_of_alc >2)].sex.count()
p = adults/children
print(f'{adults}/{children} = {p}')
```

```python
'''4
яка кількість дітей з багатодітних та малодітних сімей
(famsize	розмір родини (двійковий: "LE3" - менше або дорівнює 3 або "GT3" - більше 3)) 
споживають багато алкоголю'''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
size = data[(data.Mean_of_alc >= 3)].groupby(['famsize']).Mean_of_alc.count()
pd.DataFrame(size)
```

```python
'''5
діти, котрі хочуть вступити до вишу: за рівнем споживання алкоголю та середня к-сть пропусків
'''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
highedu = data[(data.higher == 'yes')].groupby(['higher','Mean_of_alc']).absences.mean()
pd.DataFrame(highedu)
```

```python
'''6
діти, котрі хочуть вступити до вишу: яка кількість вживає алкоголь за рівнем
та яке їх співвідношення до дітей,що не хочуть вступити до вишу за рівнями'''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
highedu1 = data[(data.higher >= 'yes')].groupby(['Mean_of_alc']).sex.count()
highedu1
```

```python
noedu = data[(data.higher >= 'no')].groupby(['Mean_of_alc']).sex.count()
noedu
```

```python
'''6
діти, котрі хочуть вступити до вишу: яка кількість вживає алкоголь за рівнем
та яке їх співвідношення до дітей,що не хочуть вступити до вишу за рівнями'''
'''порівняймо ці дві таблиці'''

data['Mean_of_alc'] = (data['Dalc']+data['Walc'])/2
highedu1 = data[(data.higher >= 'yes')].groupby(['Mean_of_alc']).sex.count()
noedu = data[(data.higher >= 'no')].groupby(['Mean_of_alc']).sex.count()

edu = pd.DataFrame(highedu1).join(pd.DataFrame(noedu),lsuffix='_high', rsuffix='_nohigh')
edu1 = edu.rename(columns = {'sex_high':'high'})
edu1.rename(columns = {'sex_nohigh':'no high'})
```
